```python
import pandas as pd
df=pd.read_csv("medical_students_dataset.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Gender"]=le.fit_transform(df["Gender"])
df["Blood Type"]=le.fit_transform(df["Blood Type"])
df["Diabetes"]=le.fit_transform(df["Diabetes"])
df["Smoking"]=le.fit_transform(df["Smoking"])
df
```

```python
df["Student ID"]=df["Student ID"].fillna(df["Student ID"].median())
df["Age"]=df["Age"].fillna(df['Age'].median())
df["Height"]=df["Height"].fillna(df["Height"].median())
df["Weight"]=df["Weight"].fillna(df["Weight"].median())
df["BMI"]=df["BMI"].fillna(df["BMI"].median())
df["Temperature"]=df["Temperature"].fillna(df["Temperature"].median())
df["Heart Rate"]=df["Heart Rate"].fillna(df["Heart Rate"].median())
df["Blood Pressure"]=df["Blood Pressure"].fillna(df["Blood Pressure"].median())
df["Cholesterol"]=df["Cholesterol"].fillna(df["Cholesterol"].median())
df
```

```python
df.isnull().sum().sum()
```

```python
from imblearn.under_sampling import RandomUnderSampler
X = df.drop(["Diabetes"], axis=1)  
y = df["Diabetes"]                
smote= RandomUnderSampler(sampling_strategy="majority")
X_sm, y_sm= smote.fit_resample(X, y)
print(X_sm.shape)
print(y_sm.shape)
```

```python
y_sm.value_counts()
```

```python
x=df.drop(["Diabetes"],axis=1)
y=df["Diabetes"]
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

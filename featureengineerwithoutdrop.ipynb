```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
df['Age']=imputer.fit_transform(df[['Age']])
df['Student ID']=imputer.fit_transform(df[['Student ID']])
df['Height']=imputer.fit_transform(df[['Height']])
df['Weight']=imputer.fit_transform(df[['Weight']])
df['BMI']=imputer.fit_transform(df[['BMI']])
df['Temperature']=imputer.fit_transform(df[['Temperature']])
df['Heart Rate ']=imputer.fit_transform(df[['Heart Rate']])
df['Blood Pressure']=imputer.fit_transform(df[['Blood Pressure']])
df['Cholesterol']=imputer.fit_transform(df[['Cholesterol']])
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy="most_frequent")
df["Gender"]=imputer.fit_transform(df[["Gender"]]).ravel()
df["Blood Type"]=imputer.fit_transform(df[["Blood Type"]]).ravel()
df["Diabetes"]=imputer.fit_transform(df[['Height']]).ravel()
df["Smoking"]=imputer.fit_transform(df[["Smoking"]]).ravel()
df.isnull().sum()
```

```python
df.drop(columns=["Heart Rate"],inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Gender"]=le.fit_transform(df["Gender"])
df["Blood Type"]=le.fit_transform(df["Blood Type"])
df["Diabetes"]=le.fit_transform(df["Diabetes"])
df["Smoking"]=le.fit_transform(df["Smoking"])
df
```

```python
x=df.drop(columns=["Smoking"],axis=1)
y=df["Smoking"]
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))

```
